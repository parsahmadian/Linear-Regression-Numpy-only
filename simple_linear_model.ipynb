# train a simple linear model using Numpy

- supervised learning
- regression

```python
import numpy as np
```

## Dataset

```python
X = np.array([
    [170, 70, 25],
    [180, 80, 30],
    [160, 60, 20]
])

y = np.array([300, 400, 250])
```

| Height | Weight | Age    | Price  |
| ------ | ------ | ------ | ------ |
| 170    | 70     | 25     | 300    |
| 180    | 80     | 30     | 400    |
| 160    | 60     | 20     | 250    |

## First Forward

```python
first_weights = np.random.randn(3) # Model weight, not the weight in the data
first_bias =  np.random.randn(1)
first_weights, first_bias
```

```python
wxn = X * first_weights
wxn
```

```python
sigma_wxn = np.sum(wxn, axis=1)
sigma_wxn
```

```python
y_pred = sigma_wxn + first_bias
y_pred
```

| y_true | y_pred |
| ------ | ------ |
| 300    | -12.41 |
| 400    | -7.69  |
| 250    | -17.14 |

## Backpropagation

### Loss Function (MSE)

```python
loss = np.sum((y - y_pred) ** 2) / len(y_pred) # MSE = sigma(y_true - y_pred)**2 / N
loss
```
